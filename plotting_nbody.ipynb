```python
fname = glob.glob('/Users/oanavesa/Desktop/GradSchool/FirstYear/ASTRO506/data_headon/data_t*.txt')

Lx_list = []
Ly_list = []
Lz_list = []
position_list =[]
velocity_list = []
vx_list = []
vy_list = []
vz_list = []

for i in range(0,len(fname)):
    xm0,ym0,zm0,vxm0,vym0,vzm0 = np.genfromtxt(fname[i], usecols=(0,1,2,3,4,5), unpack=True, skip_header=2)
    x_position = np.average(xm0)
    y_position = np.average(ym0)
    z_position = np.average(zm0)
        
    x_velocity = np.average(vxm0)
    y_velocity = np.average(vym0)
    z_velocity = np.average(vzm0)
    
    vx_list.append(x_position)
    vy_list.append(y_position)
    vz_list.append(z_position)

    arr = np.array([x_position,y_position,z_position])
    arr2 = np.array([x_velocity,y_velocity,z_velocity])
        
    Lx,Ly,Lz = np.cross(arr,arr2)
    Lx_list.append(Lx)
    Ly_list.append(Ly)
    Lz_list.append(Lz)

    
    
        
    #Lx = np.cross(x_position,(1/20000)*x_velocity)
    #Lx_list.append(Lx)
```

```python
time_array = np.linspace(0,300, 31)
```

```python
plt.figure()
plt.plot(time_array,np.array(Lx_list).transpose(), label = 'Lx')
plt.plot(time_array,np.array(Ly_list).transpose(), label = 'Ly')
plt.plot(time_array,np.array(Lz_list).transpose(), label = 'Lz')
plt.legend(shadow=True)
plt.grid()

plt.title('Angular Momentum During Merger2', fontweight='bold', fontsize='14')
plt.xlabel('Time', fontweight='bold', fontsize='12')
plt.ylabel('Angular Momentum', fontweight='bold', fontsize='12')
plt.savefig("/Users/oanavesa/Desktop/GradSchool/FirstYear/ASTRO506/project_plots/" + 'angular_momentum' + ".png", dpi=250)

plt.show()
```
