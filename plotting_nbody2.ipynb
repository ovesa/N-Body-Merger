```python
fname = glob.glob('/Users/oanavesa/Desktop/GradSchool/FirstYear/ASTRO506/data_headon/data_t*.txt')
                  #lo/data_t*.txt')

Lx_list = []
Ly_list = []
Lz_list = []
position_list =[]
velocity_list = []
vx_list = []
vy_list = []
vz_list = []

for i in range(0,len(fname)):
    xm0,ym0,zm0,vxm0,vym0,vzm0 = np.genfromtxt(fname[i], usecols=(0,1,2,3,4,5), unpack=True, skip_header=2)
    xc_sorted = [j for j in xm0 if np.logical_and(j > -20,j<20)]
    yc_sorted = [j for j in ym0 if np.logical_and(j > -10,j<10)]
    zc_sorted = [j for j in zm0 if np.logical_and(j > -10,j<10)]

    x_position = np.average(xc_sorted)
    y_position = np.average(yc_sorted)
    z_position = np.average(zc_sorted)
        
    x_velocity = np.average(vxm0)
    y_velocity = np.average(vym0)
    z_velocity = np.average(vzm0)
    
    vx_list.append(x_position)
    vy_list.append(y_position)
    vz_list.append(z_position)

    arr = np.array([x_position,y_position,z_position])
    arr2 = np.array([x_velocity,y_velocity,z_velocity])
        
    Lx,Ly,Lz = np.cross(arr,arr2)
    Lx_list.append(Lx)
    Ly_list.append(Ly)
    Lz_list.append(Lz)

    
    
        
    #Lx = np.cross(x_position,(1/20000)*x_velocity)
    #Lx_list.append(Lx)
```
